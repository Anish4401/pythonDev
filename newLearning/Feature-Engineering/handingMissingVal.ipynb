```python
import seaborn as sns
df=sns.load_dataset('titanic')
print(df.head(10))

print(df.isnull().sum())
```

```python
plot=sns.barplot(df['age'])
```

```python
df['age_mean']=df['age'].fillna(df['age'].mean())
df[['age_mean','age']]
```

```python
# Works well when we have normal distrbuted data
# 2:Median value Imputation : if we have outlier in the dataets
```

```python
df['age_median']=df['age'].fillna((df['age']).median())
df[['age_median','age_mean' ,'age','sex']]
```

<!-- Mode Imputation Techinique here we will repalce the null val with the mode of the cols values  lets gooo  -->

```python
df['age_mode']=df['age'].fillna(df['age'].mode())
df[['age_mode','age_median','age_mean','sex']]
```

```python
<!-- MOde Imputaition for : Categorical Values -->
```

```python
df[df['embarked'].isnull()]
```

```python
mode_val=df[df['embarked'].notna()]['embarked'].mode()[0]
```

```python
df['embarked_mode']=df['embarked'].fillna(mode_val)
```

```python
df[['embarked','embarked_mode']]
```
