```python
import pandas as pd 
from sklearn.preprocessing import OneHotEncoder
```

```python
# Create simple one DataFrame
df=pd.DataFrame({
    'colors':('red','blue','green','black')
})
```

```python
print(df)
```

```python
encoder=OneHotEncoder()
enconded=encoder.fit_transform(df[['colors']]).toarray()
# here it sorted in the alphabetical order and shown in the array format of 0 and 1
```

```python
encoded_df=pd.DataFrame(enconded,columns=encoder.get_feature_names_out())
```

```python
encoded_df
```

```python
encoder.transform([['red']]).toarray()
# so this fix the index for the new data and exsiting one
```

```python
import seaborn as sns
data=sns.load_dataset('tips')
data
```

```python
encodedd=OneHotEncoder()
```

```python
# OneHotEncoder does not handle multiple columns when passed as a tuple like this.
# Instead, select the columns as a DataFrame:
encoded_data = encodedd.fit_transform(data[['day', 'sex', 'smoker']]).toarray()
```

```python
encoded_data
```

```python
encoded_sns=pd.DataFrame(encoded_data,columns=encodedd.get_feature_names_out())
```
