```python
import numpy as np 
import pandas as pd 
np.random.seed(123)
sample=1000
class_ratio=0.9
n_class_1=int(sample * class_ratio)
n_class_2=(sample-n_class_1)
```

```python
n_class_1,n_class_2
```

```python
# now we have to perform the up scaling the values like match the 900 as 900,for both the class
```

```python
class_1=pd.DataFrame({
    'feature1':np.random.normal(loc=0,scale=1,size=n_class_1),
        'feature1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[0]*n_class_1

})
class_2=pd.DataFrame({
        'feature1':np.random.normal(loc=2,scale=1,size=n_class_2),
    'feature2':np.random.normal(loc=2,scale=1,size=n_class_2),
     'target':[1]*n_class_2
})
```

```python
df=pd.concat([class_1,class_2]).reset_index(drop=True)
```

```python
df.head(5)
```

```python
df['target'].value_counts()
```

```python
df_minority=df[df['target']==1]
```

```python
df_minority
```

```python
df_majority=df[df['target']==0]
```

```python
df_majority.head(5)
```

```python
# Unable to import sklearn, so let's implement a simple resampling using pandas and numpy

# Upsample the minority class (target==1) to match the majority class size
df_minority_upsampled = df_minority.sample(
    n=len(df_majority), 
    replace=True, 
    random_state=43
)
```

```python
df_minority_upsampled.shape
```

```python
df_minority_upsampled.head()
```

```python
df_upsampled=pd.concat([df_majority,df_minority_upsampled])
```

```python
df_upsampled['target'].value_counts()
# here we  got the upsacled values that is equal for both the 0 and 1 as 900 
```

```python
# Now we will downScaling to match the 100 for both 0 and 1 but the recoomeded is the upscale 
```

```python
class_1=pd.DataFrame({
    'feature1':np.random.normal(loc=0,scale=1,size=n_class_1),
        'feature1':np.random.normal(loc=0,scale=1,size=n_class_1),
    'target':[0]*n_class_1

})
class_2=pd.DataFrame({
        'feature1':np.random.normal(loc=2,scale=1,size=n_class_2),
    'feature2':np.random.normal(loc=2,scale=1,size=n_class_2),
     'target':[1]*n_class_2
})
```

```python
df=pd.concat([class_1 ,class_2,]).reset_index(drop=False)
```

```python
df
```

```python
df['target'].value_counts()
```

```python
df_point=df[df['target']==0]
```

```python
df_down=df[df['target']==1]
```

```python
df_down.head(4)
```

```python

# Downsample the majority class (class 0) to match the minority class (class 1) size
df_majority_downsampled = df_point.sample(n=len(df_down), replace=False, random_state=42)
```

```python
df_majority_downsampled.shape
```

```python
df_concat=pd.concat([df_majority_downsampled,df_down])
```
