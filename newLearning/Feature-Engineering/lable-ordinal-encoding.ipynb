```python
import seaborn as sns 
df=sns.load_dataset('tips')
```

```python
df = pd.DataFrame(df[['total_bill','time']])
```

```python
mean_price=df.groupby("time")['total_bill'].mean().to_dict()
```

```python
mean_price
```

```python
df['price']=df['time'].map(mean_price)
```
