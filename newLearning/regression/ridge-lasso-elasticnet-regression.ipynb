```python
object=[features for features in dataset.columns if dataset[features].dtype=='O']
```

```python
for i in range(6, len(object)):
    # Strip whitespace and replace problematic strings before converting to float
    dataset[object[i]] = (
        dataset[object[i]]
        .astype(str)
        .str.strip()
       .replace({'not fire': None, 'fire': None, 'nan': None, 'NaN': None, '': None})
        .astype(float)
    )
```

```python
for i in range(0,6):
    dataset[object[i]]=dataset[object[i]].astype('int')
```

```python
dataset.info()
```

```python
# 'Classes' is not in the list 'object', so we need to handle this gracefully.
# Let's check if 'Classes' is in dataset.columns and convert if present.
# 'Classes' is in the columns, but maybe the column name has extra spaces.
# Let's print the columns to debug and handle possible whitespace issues.
print("Current columns:", list(dataset.columns))
# Try to find the correct column name for 'Classes'
classes_col = None
for col in dataset.columns:
    if col.strip() == 'Classes':
        classes_col = col
        break

if classes_col:
    dataset[classes_col] = dataset[classes_col].astype('object')
else:
    print("'Classes' column not found (even after stripping whitespace).")
```

```python
df=pd.read_csv('Algerian_forest_fires_cleaned_dataset.csv',index_col=False)
```

```python
df.head()
```

```python
df_copy=df.drop(['day','month','year'],axis=1)
print(df_copy['Classes'].value_counts())

```

```python
# In pandas, the .str accessor only works with columns of string (object) dtype that actually contain string values.
# If the column contains mixed types (e.g., numbers, None, or other objects), .str will raise an error.
# By converting to string with astype(str), we ensure all values are strings, so .str.contains works safely.
df_copy['Classes'] = df_copy['Classes'].astype(str)
df_copy['Classes'] = np.where(df_copy['Classes'].str.contains('not fire'), 0, 1)
```

```python
import matplotlib.pyplot as plt
import matplotlib_inline
import seaborn as sns

# Use a valid matplotlib style, e.g., 'seaborn-v0_8' or 'ggplot'
plt.style.use('seaborn-v0_8')

# Plot histograms for all numeric columns in df_copy
plt.figure(figsize=(20, 15))
numeric_cols = df_copy.select_dtypes(include=[np.number]).columns
df_copy[numeric_cols].hist(bins=50, figsize=(20, 15), layout=(len(numeric_cols) // 3 + 1, 3))
plt.tight_layout()
plt.show()
```
