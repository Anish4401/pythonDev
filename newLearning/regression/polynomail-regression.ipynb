```python

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib_inline
```

```python
X=6*np.random.rand(100,1)-3
Y=.5 *X**2 +1.5*X +2 +np.random.rand(100,1)
plt.scatter(X,Y,)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.33, random_state=42)
print("X_train shape:", X_test.shape)
print("Y_train shape:", Y_test.shape)
```

```python
X_train
```

```python
Y_train
```

```python
from sklearn.linear_model import LinearRegression
reg1=LinearRegression()
```

```python
reg1.fit(X_train, Y_train)
```

```python
from sklearn.metrics import r2_score
res=r2_score(Y_train,reg1.predict(X_train))
print(res)
```

```python
plt.plot(X_train,reg1.predict(X_train))
plt.scatter(X_train,Y_train)
plt.xlabel("X Dataset")
plt.ylabel("Y Datset")
```

```python
# now will apply the polynomial reg  to plot this non linaer datasets 
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_2_degreee=PolynomialFeatures(degree=2,include_bias=True)
```

```python
poly_X_train=poly_2_degreee.fit_transform(X_train)
poly_X_test=poly_2_degreee.transform(X_test)
```

```python
poly_X_train
```

```python
reg1.fit(poly_X_train,Y_train)
Y_pred=reg1.predict(poly_X_test)
res_score=r2_score(Y_test,Y_pred)
print(res_score)
```

```python
plt.scatter(X_train,reg1.predict(poly_X_train))
plt.scatter(X_train,Y_train)
```

now we will use the pipeline for based on the degree provided to calx

```python
from sklearn.pipeline import Pipeline
```

```python
def poly(degree):
    X_test_new = np.linspace(-3, -3, 200).reshape(200, 1)
    new_poly = PolynomialFeatures(degree=degree, include_bias=True)
    reg_poly = LinearRegression()
    poly_reg = Pipeline([
        ("Polynomial_feature", new_poly),
        ("LinearRegression", reg_poly)
    ])
    poly_reg.fit(X_train, Y_train)
    y_pred_poly = poly_reg.predict(X_test_new)
    plt.plot(X_test_new, y_pred_poly,linewidth=3)
    plt.plot(X_train, Y_train,linewidth=3)
    plt.plot(X_test, Y_test,linewidth=3)
    plt.axis([-4,4,0,10])
    plt.show()
```

```python
poly(3)
```
